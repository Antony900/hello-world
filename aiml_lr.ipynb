```python
# linear_regression.py
import numpy as np
from sklearn.linear_model import LinearRegression

# Example data: study hours vs exam scores
X = np.array([[1], [2], [3], [4], [5]])   # study hours
y = np.array([20, 40, 60, 80, 100])       # exam scores

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict score for 6 hours of study
predicted = model.predict([[9]])
print("Predicted score for 6 hours:", predicted[0])
```
